```python
# Code For Data Generation 

import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(42)

# Number of rows (strong dataset)
rows = 5000

# Date range (randomized across full year)
dates = pd.date_range(start="2024-01-01", end="2024-12-31")

# Create dataset
data = {
    "Date": np.random.choice(dates, rows),
    "Plant_Location": np.random.choice(["Mumbai", "Pune", "Chennai"], rows),
    "Machine_ID": np.random.choice(["M1", "M2", "M3", "M4"], rows),
    "Shift": np.random.choice(["A", "B", "C"], rows),
    "Planned_Production_Time": np.random.randint(420, 480, rows),  # minutes
    "Downtime": np.random.randint(20, 120, rows),  # minutes
    "Ideal_Cycle_Time": np.random.uniform(0.8, 1.2, rows),
    "Total_Production": np.random.randint(300, 800, rows),
    "Defective_Production": np.random.randint(5, 80, rows),
    "Downtime_Reason": np.random.choice(
        ["Breakdown", "Maintenance", "Setup", "Power Failure"], rows
    )
}

df = pd.DataFrame(data)

# Derived columns
df["Operating_Time"] = df["Planned_Production_Time"] - df["Downtime"]
df["Good_Production"] = df["Total_Production"] - df["Defective_Production"]

# Ensure no negative values
df["Operating_Time"] = df["Operating_Time"].clip(lower=0)
df["Good_Production"] = df["Good_Production"].clip(lower=0)

# Save file
file_path = "../data/raw/manufacturing_data.csv"
df.to_csv(file_path, index=False)

print("✅ Dataset generated and saved successfully!")
print(f"📁 Location: {file_path}")
print(f"📊 Total Rows: {len(df)}")
```
